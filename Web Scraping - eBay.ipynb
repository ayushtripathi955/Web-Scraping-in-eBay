```python
# function to clean the values to a uniform format
def clean(val):
    if '\n' in val:
        return (val.replace('\n',''))
    elif ' ' in val:
        return (val.replace(' ',''))
    else:
        return val
```

```python
## Setting the directory to Sponsored folder and capturing all the relevant information and storing it in "details" variable to be input into the SQL Table later

os.chdir('/Users/ayushtripathi/Downloads/Sponsored')
details = []

# 1. Going to every page's file stored in the local PC and retrieving all the results of that file

for index, item in enumerate(data_sponsored_list, start = 1):
    item_no_spon = int(re.findall('/([0-9]+)',data_sponsored)[data_sponsored_list.index(item)])
    with open("%02d.htm"%item_no_spon, encoding="utf-8") as x:
        data = x.read()
        soup = BeautifulSoup(data, 'html.parser')
        x.close()

    # Store required details in a dictionary
    details_cleaned = {}
    
    ## Selector objects (for all items)
    
    if  soup.find('span',{'class':re.compile('mbg-nw')}).get_text() != None:
        details_cleaned['seller_name'] = clean(soup.find('span',{'class':re.compile('mbg-nw')}).get_text())
    else:
        details_cleaned['seller_name'] = 'None'
        
    details_cleaned['seller_score'] = clean(re.search(r"\d+",soup.find('span', {"class":re.compile("mbg-l")}).get_text()).group())
    
    if  soup.find('span',{'class':re.compile('notranslate')}).get_text() != None:
        details_cleaned['item_price'] = clean(soup.find('span',{'class':re.compile('notranslate')}).get_text()) 
    else:
        details_cleaned['item_price'] = 'None'
        
    if  soup.find('a',{'class':'vi-txt-underline'}) != None:
        details_cleaned['no_items_sold'] = clean(soup.find('a',{'class':'vi-txt-underline'})).text      
    else:
        details_cleaned['no_items_sold'] = 'None'
    
    details_cleaned['no_items_sold'] = re.sub('\W+','',details_cleaned['no_items_sold'])
    
    if  soup.find('div',{'class':'vi-bbox-dspn u-flL lable boLable'}) != None:
        details_cleaned['best_offer_available'] = 1
    else:
        details_cleaned['best_offer_available'] = 0
    
    if  soup.find('h1',{'id':re.compile('itemTitle')}).get_text() != None:
        details_cleaned['title'] = clean(soup.find('h1',{'id':re.compile('itemTitle')}).get_text())
    else:
        details_cleaned['title'] = 'None'
    
    if  soup.find('span', {"id":re.compile("vi-ret-accrd-txt")}).get_text() != None:
        details_cleaned['returns_allowed'] = clean(soup.find('span', {"id":re.compile("vi-ret-accrd-txt")}).get_text())
    else:
        details_cleaned['returns_allowed'] = 'None'
    
    if  soup.find('span',{'id':'fshippingCost'}) != None:
        details_cleaned['shipping_cost'] = soup.find('span',{'id':'fshippingCost'}).text
    else:
        details_cleaned['shipping_cost'] = 'None'
    
    details_cleaned['shipping_cost'] = re.sub('\W+','',details_cleaned['shipping_cost'])
    
    if  soup.find('div',{'id':re.compile('vi-itm-cond')}).get_text() != None:
        details_cleaned['item_condition'] = clean(soup.find('div',{'id':re.compile('vi-itm-cond')}).get_text())
    else:
        details_cleaned['item_condition'] = 'None'
    
    details_cleaned['Sponsored'] = 1
    
    details.append(details_cleaned)
    
time.sleep(3)
```
